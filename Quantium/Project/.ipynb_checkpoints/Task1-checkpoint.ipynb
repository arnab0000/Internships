```python
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt
purchase_behaviour = pd.read_csv('QVI_purchase_behaviour.csv')
pb_df = purchase_behaviour.copy()
pb_df.dropna(inplace=True)
transaction_data = pd.read_excel('QVI_transaction_data.xlsx')
td_df = transaction_data.copy()
td_df.dropna(inplace=True)
result_df = pd.merge(td_df, pb_df, on=['LYLTY_CARD_NBR'])
print(result_df)
```

```python
result_df
```

```python
result_df['PACK_SIZE'] = result_df['PROD_NAME'].str.extract('(\d+)')
```

```python
#removing special characters
import re
result_df['PROD_NAME'] = result_df['PROD_NAME'].map(lambda x: re.sub(r'\W+', '', x))
```

```python
#removing the numerical part of the sting
import string
result_df['PROD_NAME'] = result_df['PROD_NAME'].str.replace('\d+', "")
result_df['PROD_NAME'] = result_df['PROD_NAME'].apply(lambda x: x[:-1].strip())
result_df
```

```python
df = result_df[~result_df["PROD_NAME"].str.contains("Salsa")]
df = df[~df["PROD_NAME"].str.contains("SALSA")]
df = df[~df["PROD_NAME"].str.contains("Salsa")]
df
```

```python
df['PROD_NAME'].value_counts()
```

```python
df.dropna(inplace=True)
```

```python
df[df["PROD_QTY"] >= 200]
```

```python
df[df['LYLTY_CARD_NBR'] == 226000]
```

```python
#converting 5 digit numbers to date
#these 5 digits number are days and csv date starts from 30-12-1899. Now when we convert this date to a 5 digit number we get 25569. we use this number as origin
df['DATE'] = pd.to_datetime((df['DATE'] - 25569) * 86400.0, unit='s')
```
